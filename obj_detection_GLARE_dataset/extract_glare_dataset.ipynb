```python
for vid_folder in os.listdir(root_folder):
    vid_path_temp = os.path.join(root_folder, vid_folder)
    # print(vid_path_temp)
    # annotation_file = os.path.join(vid_path_temp, 'frameAnnotations.csv')
    # print(os.path.isfile(annotation_file))

    # print(os.path.isdir(vid_path_temp))
    if os.path.isdir(vid_path_temp):

        for vid_folder2 in os.listdir(vid_path_temp): 
            vid_path = os.path.join(vid_path_temp, vid_folder2)
            print(vid_path)
            annotation_file = os.path.join(vid_path, 'frameAnnotations.csv')
            print(annotation_file)
            print(f"isFile: {os.path.isfile(annotation_file)}")
            # print(os.path.isdir(annotation_file))
            annotations = pd.read_csv(annotation_file)
            for idx, row in annotations.iterrows():
                image_name = row['Filename']  # Adjust file extension if different
                annotation_tag = row['Annotation tag']

                # Define bounding box coordinates
                upper_left_x = row['Upper left corner X']
                upper_left_y = row['Upper left corner Y']
                lower_right_x = row['Lower right corner X']
                lower_right_y = row['Lower right corner Y']

                # Create the folder for the annotation tag if it doesn't exist
                tag_folder = os.path.join(output_folder, annotation_tag)
                os.makedirs(tag_folder, exist_ok=True)

                image_path = os.path.join(vid_path, image_name)

                if os.path.isfile(image_path):
                    with Image.open(image_path) as img:
                        cropped_img = img.crop((upper_left_x, upper_left_y, lower_right_x, lower_right_y))
                        
                        # Save the cropped image in the tag folder
                        cropped_image_path = os.path.join(tag_folder, f"{vid_folder}_{idx}_{image_name}")
                        cropped_img.save(cropped_image_path)
                        print(f"Saved cropped image: {cropped_image_path}")

                else:
                    print(f"Image not found: {image_path}")

                   


```

```python

import shutil
import random
```

```python
# Paths for input and output
input_folder = 'extracted_glare_dataset'  # Folder with the cropped images organized by tag
output_folder = 'yolo_style_dataset'  # Folder where the new structure will be created
train_ratio = 0.8 
```

```python
images_train_folder = os.path.join(output_folder, 'images', 'train')
images_val_folder = os.path.join(output_folder, 'images', 'val')
labels_train_folder = os.path.join(output_folder, 'labels', 'train')
labels_val_folder = os.path.join(output_folder, 'labels', 'val')
```

```python
# Create directories if they don't exist
os.makedirs(images_train_folder, exist_ok=True)
os.makedirs(images_val_folder, exist_ok=True)
os.makedirs(labels_train_folder, exist_ok=True)
os.makedirs(labels_val_folder, exist_ok=True)
```

```python
# Process each tag folder in extracted_glare_dataset
for tag_folder in os.listdir(input_folder):
    tag_path = os.path.join(input_folder, tag_folder)
    
    if os.path.isdir(tag_path):
        # Get all images in the tag folder
        images = [f for f in os.listdir(tag_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
        
        # Shuffle and split images into train and val sets
        random.shuffle(images)
        split_index = int(len(images) * train_ratio)
        train_images = images[:split_index]
        val_images = images[split_index:]
        
        for image_name in train_images:
            # Move image to the train images folder
            image_path = os.path.join(tag_path, image_name)
            shutil.copy(image_path, os.path.join(images_train_folder, image_name))
            
            # Assume labels are in the same format with .txt extension
            label_name = image_name.rsplit('.', 1)[0] + '.txt'
            print(label_name)
            label_path = os.path.join(tag_path, label_name)
            if os.path.exists(label_path):
                shutil.copy(label_path, os.path.join(labels_train_folder, label_name))
        
        for image_name in val_images:
            # Move image to the val images folder
            image_path = os.path.join(tag_path, image_name)
            shutil.copy(image_path, os.path.join(images_val_folder, image_name))
            
            # Assume labels are in the same format with .txt extension
            label_name = image_name.rsplit('.', 1)[0] + '.txt'
            label_path = os.path.join(tag_path, label_name)
            if os.path.exists(label_path):
                shutil.copy(label_path, os.path.join(labels_val_folder, label_name))

print("Dataset split and organized successfully.")
```
