```python
from ultralytics import YOLO

# Load a YOLO11 model (choose the appropriate variant: n, s, m, l, x)
model = YOLO('yolov8n.pt')  # 'n' denotes the nano version; adjust as needed

# Train the model
model.train(data='yolo_glare_dataset/data.yaml', epochs=100, imgsz=640)
```

```python

# Save the trained model
model.save('saved_yolo_glare_model2.pt')
```

```python
import cv2
from ultralytics import YOLO

# Load the fine-tuned YOLOv8 model
model = YOLO('saved_yolo_glare_model2.pt')  # Path to the trained model weights

# Open a connection to the webcam
camera = cv2.VideoCapture(0)

while camera.isOpened():
    success, frame = camera.read()
    if not success:
        break

    # Perform inference
    results = model(frame,conf=0.3)

    # # Extract and draw bounding boxes for detected objects
    # for detection in results[0].boxes:  # Each detection
    #     x1, y1, x2, y2, confidence, class_id = detection
    #     cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)  # Draw bounding box

    # Extract and draw bounding boxes for detected objects
    for box in results[0].boxes:  # Each detected bounding box
        x1, y1, x2, y2 = box.xyxy[0]  # Get box coordinates
        confidence = box.conf[0]      # Get confidence score
        class_id = box.cls[0]         # Get class ID

        # Draw bounding box on frame
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)


    # Display frame
    cv2.imshow("Traffic Sign Detection", frame)

    # Break on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
camera.release()
cv2.destroyAllWindows()
```
